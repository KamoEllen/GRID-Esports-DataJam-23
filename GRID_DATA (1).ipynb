```python
import pandas as pd
from tabulate import tabulate

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    #"2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Columns to select
columns_to_select = ["name", "score", "kills", "killAssistsReceived", "killAssistsGiven"]

# Dictionary to store team statistics
team_statistics = {}

# Read and process each CSV file
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url, usecols=columns_to_select)
    
    for index, row in df.iterrows():
        team_name = row["name"]
        if team_name not in team_statistics:
            team_statistics[team_name] = {
                "total_matches": 0,
                "total_score": 0,
                "total_kills": 0,
                "total_kill_assists_received": 0,
                "total_kill_assists_given": 0
            }
        team_statistics[team_name]["total_matches"] += 1
        team_statistics[team_name]["total_score"] += row["score"]
        team_statistics[team_name]["total_kills"] += row["kills"]
        team_statistics[team_name]["total_kill_assists_received"] += row["killAssistsReceived"]
        team_statistics[team_name]["total_kill_assists_given"] += row["killAssistsGiven"]

# Create a list to store rows of the table
table_rows = []

# Analyze statistics and add rows to the table_rows list
for team_name, stats in team_statistics.items():
    total_matches = stats["total_matches"]
    
    avg_score = stats["total_score"] / total_matches
    avg_kills = stats["total_kills"] / total_matches
    avg_kill_assists_received = stats["total_kill_assists_received"] / total_matches
    avg_kill_assists_given = stats["total_kill_assists_given"] / total_matches
    
    row = [team_name, f"{avg_score:.2f}", f"{avg_kills:.2f}", f"{avg_kill_assists_received:.2f}", f"{avg_kill_assists_given:.2f}"]
    table_rows.append(row)

# Create a DataFrame using the table_rows list and define column names
columns = ["Team", "Avg. Score", "Avg. Kills", "Avg. Kill Assists Received", "Avg. Kill Assists Given"]
df_table = pd.DataFrame(table_rows, columns=columns)

# Format the DataFrame as a table using the tabulate function
table_avg = tabulate(df_table, headers='keys', tablefmt='pretty', showindex=False)

# Display the formatted table
print(table_avg)
```

```python
import pandas as pd
from tabulate import tabulate

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    #"2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Columns to select
columns_to_select = ["name", "score", "kills", "killAssistsReceived", "killAssistsGiven", "won"]

# Dictionary to store team statistics
team_statistics = {}

# Read and process each CSV file
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url, usecols=columns_to_select)
    
    for index, row in df.iterrows():
        team_name = row["name"]
        if team_name not in team_statistics:
            team_statistics[team_name] = {
                "total_matches": 0,
                "total_score": 0,
                "total_kills": 0,
                "total_wins": 0
            }
        team_statistics[team_name]["total_matches"] += 1
        team_statistics[team_name]["total_score"] += row["score"]
        team_statistics[team_name]["total_kills"] += row["kills"]

        if row["won"]:
            team_statistics[team_name]["total_wins"] += 1


# Identify top performers and create tables
top_performers = {
    "score": [],
    "kills": []
}

for team_name, stats in team_statistics.items():
    total_matches = stats["total_matches"]
    if total_matches > 0:
        avg_score = stats["total_score"] / total_matches
        avg_kills = stats["total_kills"] / total_matches

        top_performers["score"].append((team_name, avg_score))
        top_performers["kills"].append((team_name, avg_kills))

# Function to create a formatted table for top performers
def create_top_performers_table(metric_name, metric_list):
    sorted_list = sorted(metric_list, key=lambda x: x[1], reverse=True)
    headers = ["Rank", "Team", f"Avg. {metric_name}"]
    rows = []
    for idx, (team_name, value) in enumerate(sorted_list):
        row = [idx + 1, team_name, f"{value:.2f}"]
        rows.append(row)
    return tabulate(rows, headers=headers, tablefmt='pretty')

# Create formatted tables for top performers for each metric
table_score = create_top_performers_table("Score", top_performers["score"])
table_kills = create_top_performers_table("Kills", top_performers["kills"])

# Display the formatted tables for top performers
print("Top Performers for Score:")
print(table_score)
print("\nTop Performers for Kills:")
print(table_kills)
```

```python
import pandas as pd
from tabulate import tabulate

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    #"2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Columns to select
columns_to_select = ["name", "score", "kills", "killAssistsReceived", "killAssistsGiven", "won"]

# Dictionary to store team statistics
team_statistics = {}

# Read and process each CSV file
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url, usecols=columns_to_select)
    
    for index, row in df.iterrows():
        team_name = row["name"]
        if team_name not in team_statistics:
            team_statistics[team_name] = {
                "total_matches": 0,
                "total_score": 0,
                "total_kills": 0,
                "total_kill_assists_received": 0,
                "total_kill_assists_given": 0,
                "total_wins": 0
            }
        team_statistics[team_name]["total_matches"] += 1
        team_statistics[team_name]["total_score"] += row["score"]
        team_statistics[team_name]["total_kills"] += row["kills"]
        team_statistics[team_name]["total_kill_assists_received"] += row["killAssistsReceived"]
        team_statistics[team_name]["total_kill_assists_given"] += row["killAssistsGiven"]
        if row["won"]:
            team_statistics[team_name]["total_wins"] += 1

# Create a DataFrame with relevant data
data = {
    "score": [stats["total_score"] / stats["total_matches"] for stats in team_statistics.values()],
    "kills": [stats["total_kills"] / stats["total_matches"] for stats in team_statistics.values()],
    "kill_assists_received": [stats["total_kill_assists_received"] / stats["total_matches"] for stats in team_statistics.values()],
    "kill_assists_given": [stats["total_kill_assists_given"] / stats["total_matches"] for stats in team_statistics.values()]
}
df = pd.DataFrame(data, index=team_statistics.keys())

# Calculate correlations
correlation_matrix = df.corr()

# Identify patterns in the correlation matrix and create a table
correlation_rows = []

for col1 in correlation_matrix.columns:
    for col2 in correlation_matrix.columns:
        if col1 != col2:
            correlation_value = correlation_matrix.loc[col1, col2]
            if correlation_value > 0.5 or correlation_value < -0.5:
                row = [col1, col2, f"{correlation_value:.2f}"]
                correlation_rows.append(row)

# Create a DataFrame for correlations using the correlation_rows list and define column names
correlation_columns = ["Metric 1", "Metric 2", "Correlation"]
df_correlation = pd.DataFrame(correlation_rows, columns=correlation_columns)

# Format the DataFrame as a table using the tabulate function
correlation_table = tabulate(df_correlation, headers='keys', tablefmt='pretty', showindex=False)

# Display the formatted table
print("Patterns in the Correlation Matrix:")
print(correlation_table)
```

```python
import pandas as pd
from collections import Counter

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    "2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Initialize a Counter to track column names
column_counter = Counter()

# Read and count column names for each CSV file
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url)
    column_names = df.columns.tolist()  # Get the list of column names
    column_counter.update(column_names)  # Update the counter with column names

# Create tables for different frequency counts
table_template = "{:<40} {:<10}"

print("Column Names by Frequency:")
print(table_template.format("Column Name", "Count"))
print("-" * 50)

# Group column names by frequency
grouped_by_frequency = {}
for column_name, count in column_counter.items():
    if count not in grouped_by_frequency:
        grouped_by_frequency[count] = []
    grouped_by_frequency[count].append(column_name)

# Create tables for each frequency group
for frequency, column_names in grouped_by_frequency.items():
    print(f"Column Names that show up {frequency} times:")
    print(table_template.format("Column Name", "Count"))
    print("-" * 50)
    for column_name in column_names:
        print(table_template.format(column_name, frequency))
    print("\n")
```

```python
import pandas as pd
from collections import Counter

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    "2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Initialize a Counter to track weapon types and kills
weapon_counter = Counter()

# Read and count kills for each weapon type
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url)
    weapon_columns = [col for col in df.columns if col.startswith("players/") and "/weaponKills/" in col]
    for col in weapon_columns:
        weapon_type = col.split("/weaponKills/")[-1]
        weapon_kills = df[col].sum()
        weapon_counter[weapon_type] += weapon_kills

# Find most and least kills
most_common_weapon = weapon_counter.most_common(1)[0]
least_common_weapon = weapon_counter.most_common()[-1]

# Display results
print("Number of Weapon Types:", len(weapon_counter))
print("Most Kills:", most_common_weapon)
print("Least Kills:", least_common_weapon)
```
