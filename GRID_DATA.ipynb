```python
import pandas as pd

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    #"2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Columns to select
columns_to_select = ["name", "score", "kills", "killAssistsReceived", "killAssistsGiven"]

# Dictionary to store team statistics
team_statistics = {}

# Read and process each CSV file
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url, usecols=columns_to_select)
    
    for index, row in df.iterrows():
        team_name = row["name"]
        if team_name not in team_statistics:
            team_statistics[team_name] = {
                "total_matches": 0,
                "total_score": 0,
                "total_kills": 0,
                "total_kill_assists_received": 0,
                "total_kill_assists_given": 0
            }
        team_statistics[team_name]["total_matches"] += 1
        team_statistics[team_name]["total_score"] += row["score"]
        team_statistics[team_name]["total_kills"] += row["kills"]
        team_statistics[team_name]["total_kill_assists_received"] += row["killAssistsReceived"]
        team_statistics[team_name]["total_kill_assists_given"] += row["killAssistsGiven"]

# Calculate average statistics and print the results
print("Team Statistics:")
for team_name, stats in team_statistics.items():
    avg_score = stats["total_score"] / stats["total_matches"]
    avg_kills = stats["total_kills"] / stats["total_matches"]
    avg_kill_assists_received = stats["total_kill_assists_received"] / stats["total_matches"]
    avg_kill_assists_given = stats["total_kill_assists_given"] / stats["total_matches"]
    
    print(f"Team: {team_name}")
    print(f"Average Score: {avg_score:.2f}")
    print(f"Average Kills: {avg_kills:.2f}")
    print(f"Average Kill Assists Received: {avg_kill_assists_received:.2f}")
    print(f"Average Kill Assists Given: {avg_kill_assists_given:.2f}")
    print()
```

```python
import pandas as pd

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    #"2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Columns to select
columns_to_select = ["name", "score", "kills", "killAssistsReceived", "killAssistsGiven", "won"]

# Dictionary to store team statistics
team_statistics = {}

# Read and process each CSV file
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url, usecols=columns_to_select)
    
    for index, row in df.iterrows():
        team_name = row["name"]
        if team_name not in team_statistics:
            team_statistics[team_name] = {
                "total_matches": 0,
                "total_score": 0,
                "total_kills": 0,
                "total_kill_assists_received": 0,
                "total_kill_assists_given": 0,
                "total_wins": 0
            }
        team_statistics[team_name]["total_matches"] += 1
        team_statistics[team_name]["total_score"] += row["score"]
        team_statistics[team_name]["total_kills"] += row["kills"]
        team_statistics[team_name]["total_kill_assists_received"] += row["killAssistsReceived"]
        team_statistics[team_name]["total_kill_assists_given"] += row["killAssistsGiven"]
        if row["won"]:
            team_statistics[team_name]["total_wins"] += 1

# Identify top performers and print the results
print("Top Performers:")
top_performers = {
    "score": [],
    "kills": [],
    "kill_assists_received": [],
    "kill_assists_given": []
}

for team_name, stats in team_statistics.items():
    total_matches = stats["total_matches"]
    if total_matches > 0:
        avg_score = stats["total_score"] / total_matches
        avg_kills = stats["total_kills"] / total_matches
        avg_kill_assists_received = stats["total_kill_assists_received"] / total_matches
        avg_kill_assists_given = stats["total_kill_assists_given"] / total_matches
        
        top_performers["score"].append((team_name, avg_score))
        top_performers["kills"].append((team_name, avg_kills))
        top_performers["kill_assists_received"].append((team_name, avg_kill_assists_received))
        top_performers["kill_assists_given"].append((team_name, avg_kill_assists_given))

# Function to print top performers for a given metric
def print_top_performers(metric_name, metric_list):
    sorted_list = sorted(metric_list, key=lambda x: x[1], reverse=True)
    print(f"Top Performers for {metric_name}:")
    for idx, (team_name, value) in enumerate(sorted_list):
        print(f"{idx + 1}. {team_name}: {value:.2f}")

print_top_performers("Score", top_performers["score"])
print_top_performers("Kills", top_performers["kills"])
print_top_performers("Kill Assists Received", top_performers["kill_assists_received"])
print_top_performers("Kill Assists Given", top_performers["kill_assists_given"])
```

```python
import pandas as pd

# Base URL for the GitHub repository
base_url = "https://github.com/KamoEllen/GRID-Esports-DataJam-23/raw/main/"

# List of file names
file_names = [
    "1597139_state.csv",
    "1598060_state.csv",
    "1614507_state.csv",
    "2432453_state.csv",
    "2432302_state.csv",
    #"2364966_state.csv",
    "2579089_state.csv",
    "2578928_state.csv",
    "2579048_state.csv"
]

# Columns to select
columns_to_select = ["name", "won"]

# Dictionary to store team statistics
team_statistics = {}

# Read and process each CSV file
for file_name in file_names:
    file_url = base_url + file_name
    df = pd.read_csv(file_url, usecols=columns_to_select)
    
    for index, row in df.iterrows():
        team_name = row["name"]
        if team_name not in team_statistics:
            team_statistics[team_name] = {
                "total_matches": 0,
                "total_wins": 0
            }
        team_statistics[team_name]["total_matches"] += 1
        if row["won"]:
            team_statistics[team_name]["total_wins"] += 1

# Calculate win-loss ratio for each team
win_loss_ratios = {}
for team_name, stats in team_statistics.items():
    total_wins = stats["total_wins"]
    total_matches = stats["total_matches"]
    if total_matches > 0:
        win_loss_ratio = total_wins / total_matches
        win_loss_ratios[team_name] = win_loss_ratio

# Sort teams based on win-loss ratio and print the results
sorted_teams = sorted(win_loss_ratios.items(), key=lambda x: x[1], reverse=True)
print("Win-Loss Ratios:")
for idx, (team_name, ratio) in enumerate(sorted_teams):
    print(f"{idx + 1}. {team_name}: Win-Loss Ratio = {ratio:.2f}")
```
